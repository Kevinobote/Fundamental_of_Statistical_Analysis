## Descriptive Statistics Activity 1.1


To represent the data for the number of litres of petrol purchased by customers at a petrol filling station, we'll create a pie chart, histogram, and frequency polygon.

### Given Data
| Number of litres of petrol purchased | Number of customers |
|-------------------------------------|---------------------|
| 0 and < 2                           | 15                  |
| 2 and < 4                           | 40                  |
| 4 and < 6                           | 65                  |
| 6 and < 8                           | 40                  |
| 8 and < 10                          | 30                  |
| 10 and < 12                         | 10                  |


```python

import matplotlib.pyplot as plt

# Data
intervals = ['0-2', '2-4', '4-6', '6-8', '8-10', '10-12']
customers = [15, 40, 65, 40, 30, 10]
total_customers = sum(customers)
print(total_customers)
```

### Pie Chart
A pie chart shows the proportion of each category relative to the whole. To create a pie chart, we first need to calculate the percentage of total customers in each category.


Total number of customers = 200

| Number of litres | Number of customers | Percentage |
|------------------|---------------------|------------|
| 0 and < 2        | 15                  | 7.5%       |
| 2 and < 4        | 40                  | 20%        |
| 4 and < 6        | 65                  | 32.5%      |
| 6 and < 8        | 40                  | 20%        |
| 8 and < 10       | 30                  | 15%        |
| 10 and < 12      | 10                  | 5%         |

To create the pie chart, each slice corresponds to the percentage of customers in each interval.

```python
# Pie Chart
percentages = [x / total_customers * 100 for x in customers]
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=intervals, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Petrol Purchased')
plt.show()
```

### Histogram
A histogram represents the frequency of data points in successive intervals. In this case, the intervals are the ranges of litres purchased.
To create the histogram, plot the number of litres of petrol purchased on the x-axis and the number of customers on the y-axis. Each bar represents the number of customers within each interval.

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.bar(intervals, customers, width=0.8, color='skyblue')
plt.xlabel('Number of Litres of Petrol Purchased')
plt.ylabel('Number of Customers')
plt.title('Histogram of Petrol Purchased')
plt.show()
```

### Frequency Polygon
A frequency polygon is a graphical representation of the frequencies at the midpoints of the intervals.
To create the frequency polygon, plot the midpoints of each interval on the x-axis and the number of customers on the y-axis, then connect the points with straight lines.



```python
# Frequency Polygon
midpoints = [1, 3, 5, 7, 9, 11]
plt.figure(figsize=(8, 6))
plt.plot(midpoints, customers, marker='o', linestyle='-', color='blue')
plt.xlabel('Midpoint of Number of Litres of Petrol Purchased')
plt.ylabel('Number of Customers')
plt.title('Frequency Polygon of Petrol Purchased')
plt.grid(True)
plt.show()
```
